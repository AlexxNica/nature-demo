## Interactive illustration of aliasing

Notebooks can also link code and data to user interfaces (sliders, checkboxes, dropdowns) that enable interactive exploration.

Aron Ahmadia (ERDC-CHL) and David Ketcheson (KAUST) created the following example to illustrate the perils of aliasing. Aliasing occurs when a rapidly-changing periodic signal is sampled too infrequently, which creates a faulty impression of the true frequency of the signal. 

Ketcheson explains:

> As an undergraduate, I did some observational astronomy looking at variable stars.
> These are stars whose brightness oscillates, usually on a fairly regular basis. 
> Many published results claim to measure how quickly the star's brightness oscillates - 
> but actually report the oscillations at some multiple of the real answer, due to
> insufficient observation and (as a result) aliasing.

This example shows how trying to reconstruct a simple sine wave signal from discrete measurements can fail. The sliders allow you to adjust the frequency of the underlying periodic sine wave signal (represented by the `wavenumber`), and also how often the signal is sampled (represented by the number of discrete grid points, `grid_points`, at which the amplitude of the underlying signal is measured). Get it wrong, and a high-frequency sine wave is measured as a lower-frequency signal.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from IPython.html.widgets import interact

def plot_sine(wavenumber=4.0, grid_points=12, plot_exact=True):
    "Plot sin(2*pi*p), sampled at m equispaced points."
    x  = np.linspace(0,1, grid_points+2);
    xf = np.linspace(0, 1, 1000)
    y  = np.sin(wavenumber*np.pi*x)
    yf = np.sin(wavenumber*np.pi*xf)
    fig = plt.figure(figsize=(8, 6));
    ax = fig.add_subplot(1,1,1);
    plt.xlabel('x')
    plt.ylabel('signal')
    plt.title('Aliasing in discretely sampled periodic signal')
    if plot_exact:
        ax.plot(xf, yf, 'r-', linewidth=2);
    ax.plot(x,  y,  'o-', lw=2)

interact(plot_sine, wavenumber=(1.0, 44.0), grid_points=(10, 16, 1), plot_exact=True);
```

To see the effects of aliasing:

1. Run the previous cell, then set the `grid_points` slider to `13`.
2. Increase the `wavenumber` slider to values above `20`.
3. As the wavenumber increases, the measured signal (blue) has a lower frequency than the real one (red).

## Counting stars in the Hubble deep field

Using Determinant of Hessian

```python
%matplotlib inline
from matplotlib import pyplot as plt
from skimage import data
from skimage.feature import blob_doh
from skimage.color import rgb2gray
from math import sqrt
from IPython.html.widgets import interact, fixed

image = data.hubble_deep_field()[0:500, 0:500]
image_gray = rgb2gray(image)

def find_blobs(i, max_sigma=30, threshold=0.1):
    blobs = blob_doh(i, max_sigma=max_sigma, threshold=threshold)
    return blobs

def plot_blobs(i, max_sigma=30, threshold=0.1):
    blobs = find_blobs(i, max_sigma=max_sigma, threshold=threshold)
    fig, ax = plt.subplots(1, 1, figsize=(10,0))
    ax.set_title('Counting Stars in the Hubble Deep Field')
    ax.imshow(image, interpolation='nearest')
    for blob in blobs:
        y, x, r = blob
        c = plt.Circle((x, y), r, color='yellow', linewidth=2, fill=False)
        ax.add_patch(c)

interact(plot_blobs, i=fixed(image_gray), max_sigma=(10,40,2), threshold=(0.005,0.02,0.001));
```
